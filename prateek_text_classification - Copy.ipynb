```python
# this function is used to send a list of all the words present in the textfile recieved
# the list of words returned will not contain any special characters or stop words
# file_in_txt is the file in txt mode of which the list is to be returned
def words_to_use(file_in_txt):
#   in this loop all the special characters are replaced with blank spaces
    for i in special_characters:
        file_in_txt=file_in_txt.replace(i," ")
#   all our stopwords are in lower case so converting the txt file to lower case
    file_in_txt=file_in_txt.lower()
#   list_of_words will contain all the words in txt file(with stopwords)
    list_of_words=file_in_txt.split()
#   new_list is the final list to be returned
#   in the loop below all those words are added to new_list which are not stopwords
    new_list=[]
    for i in list_of_words:
        if(i not in stop_words):
            new_list.append(i)
#   the final list is send
    return new_list
```

```python
# final_list is the list which will contains many lists
# each list of final_list contains the all the words(without stopwords) in the corresponding file in the all_files
final_list=[]
# j is just a counter to know the progress while executing the code 
# counters throughout the code are just so that the user can know how much part of the code had been executed
# j will be printed each time the loop executes, informing the no. of files read 
j=1
# in the loop below all the files will be read one by one and sent to the function words_to_use
# the list returned by the funtion will be added to the final_list
for i in all_files:
    print(j)
    j+=1
    file=open(i)
    file_in_txt=file.read()
    final_list.append(words_to_use(file_in_txt))
    file.close()
```

```python
# in this cell a new dataframe ndf will be formed which will contain only usefull features
# all_columns will be a dictionary wich will contain features as keys and the sum of all the times that feature has appeared
all_columns={}
for i in df:
    if(i==".folder."):
        continue
    all_columns[i]=df[i].sum()
# sorted_columns will be the list of tupples obtained after sorting the dictionary all_columns on the basis of values
sorted_columns = sorted(all_columns.items(), key=lambda kv: kv[1])
# the sorted_columns will be reversed to get the values in descending order
sorted_columns.reverse()
# useless_columns will be the list containing column or feature names which appear very less 
useless_columns=[]
# only the first 5000 features will considered usefull
for i in range(5000,len(sorted_columns)):
    useless_columns.append(sorted_columns[i][0])
# a copy of original dataframe df will be made and the useless_columns will be dropped
# ndf will be our final dataset to be used
ndf=df.copy()
ndf.drop(useless_columns,axis=1,inplace=True)
ndf
```

```python
# imported tosplit the ndf into testing and training data 
from sklearn import model_selection
# X will contain all features
# y will contain .folder.
X=ndf.iloc[:,:-1]
Y=ndf.iloc[:,-1]
# x_train and y_train are training data
# x_test and y_test are testing data
x_train,x_test,y_train,y_test = model_selection.train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
# this function is used to make a dictionary count_dictionary
# this dictionary will have the group names as the key and a key "total" which contains total number of words
# each of these keys will have the value as another dictionary which will have the keys of the feature names and a "total"
# the total will tell the total no of words in all the text files of that group and the keys as feature names will tell the
# number of times that feature had occured in the textof that group only
def calculate_count(x_train,y_train):
    count_dictionary={}
#   total_words=total no. of words in all the files of in the groups
    total_words=0
#   i is again an counter to tell the data of how many groups have been added 
    i=0
#   current_class will take the value as each group name
    for current_class in set(y_train):
        print(i)
        i+=1
#       a dictionary will be created for each class
        count_dictionary[current_class]={}
#       x will be the data of only those files which belong to current_class 
        x=x_train[y_train==current_class]
#       class_sum = no. of total words of all the files of the current_class 
        class_sum=0
#       current_feature will take the name of each feature one by one
        for current_feature in x_train:
#           values added to the dictionary as explained
            count_dictionary[current_class][current_feature]=x[current_feature].sum()
#           class_sum calculated
            class_sum+=count_dictionary[current_class][current_feature]
#       class_sum saved
        count_dictionary[current_class]["total"]=class_sum
#       total_words calculated
        total_words+=class_sum
#   total_words saved
    count_dictionary["total"]=total_words
#   count_dictionary returned
    return count_dictionary
```

```python
# the fit function will call the count_dictionary funtion and get count_dictionary from it to for a prob_dictionary
# this dictionary will have the group names as the key and a key "class_probability" which contains probabilty of the class
# each of these keys will have the value as another dictionary which will have the keys of the feature names 
# the keys as feature names will tell the probability that feature had occured in the text of that group only
def fit(x_train,y_train):
#   count_dictionary obtained by calling function calculate_count
    count_dictionary=calculate_count(x_train,y_train)
#   prob_dictionary formed
    prob_dictionary={}
#   i is again an counter to tell the no of classes whose data has been entered in the prob_dictionary
    i=0
#   current_class will take the name of each group one by one
    for current_class in set(y_train):
        print(i)
        i+=1
#       a dictionary will be created for each class
        prob_dictionary[current_class]={}
#       current_feature will take the name of each feature one by one
        for current_feature in x_train:
#           calculating the probability for each feature
            num=count_dictionary[current_class][current_feature]+1
            denom=count_dictionary[current_class]["total"]+len(count_dictionary[current_class].keys())
            prob_dictionary[current_class][current_feature]=m.log(num)-m.log(denom)
#       calculating the probabilty of each class(group)
        num=count_dictionary[current_class]["total"]+1
        denom=count_dictionary["total"]+len(count_dictionary.keys())
        prob_dictionary[current_class]["class_probability"]=m.log(num)-m.log(denom)
#   returning the prob_dictionary
    return prob_dictionary
```

```python
# calling fit function and getting the dictionary with probabilities
dictionary=fit(x_train,y_train)
```

```python
# this funtion will calculate the probability of x to be of the current_class using dictionary
def probability(dictionary,x,current_class):
#   output will be initialised with the probability of the current_class
    output=dictionary[current_class]["class_probability"]
    all_features=x.keys()
#   the probability of each function will be added in the given manner
    for i in range(len(x)):
        current_feature=all_features[i]
        output=output+x[i]*dictionary[current_class][current_feature]
#   output calculated will be returned
    return output
```

```python
# this function receives the x .i.e a row of the testing data and predicts which group it belongs to
def predict_for_x(dictionary,x):
#   best_p = best probability or highest probability
#   best_class = class(group) with best_p as its probability
    best_p=-1000
#   first is true for now, it is used so that the best_p can take a value in the first iteration of the loop
    first=True
    best_class=""
#   current_class will take the name of each group one by one
    for current_class in dictionary:
#       current_class_probability is the probabilty of this class to be a possible result
#       it will be calculated  by the function probability
        current_class_probability=probability(dictionary,x,current_class)
#       if current_class_probability is greater than the best_p then best_p and best_class will be updated
        if(first or current_class_probability>best_p):
            best_p=current_class_probability
            best_class=current_class
            first=False
#   returning the best_class
    return best_class
```

```python
# this function is used to return a list of predicted groups for each file
def predict(dictionary,x_test):
#   y_pred is the list which will contain all the predicted values of groups
    y_pred=[]
#   for each row of testing data the dictionary and that row is sent as parameters to the function predict_for_x
#   the function predict_for x return the predicted value which stored in the list y_pred
    for x in range(len(x_test)):
        y_pred.append(predict_for_x(dictionary,x_test.iloc[x,:]))
#       this is printed so as to tell how many groups have been predicted while running the code(this is another counter) 
        print(x,y_pred[-1])
#   y_pred is returned 
    return y_pred
```

```python
# predict called to get the predictions of groups
y_pred=predict(dictionary,x_test)
```

```python
# calculates and returns the score of result of own implementation of multinomial naive bayes
def myscore(y_pred,y_test):
    output=(y_pred==y_test).sum()
    output=output/len(y_pred)
    return output
```
